```python
#!pip3 install jieba
import pickle
import pandas as pd
import numpy as np
import csv
import jieba
import matplotlib.pyplot as plt
```

## 匯入要進行分析的文檔 (3868篇)
### - 新聞橘報 (搜尋關鍵字"電商"的結果)
### - 科技新報 (搜尋關鍵字"電商"的結果)

```python
#讀取兩份文檔: techorange 與 technews
techorange = pickle.load(open("tech_orange.pkl", 'rb'), encoding='utf-8')

with open('technews.csv', newline='',encoding = 'utf8') as csvfile:
    technews = csv.reader(csvfile)
    b = list(technews)
    #for row in technews:
        #print(row)
```

```python
# 轉乘dataframe並合併兩份文檔
technews_df = pd.DataFrame(b)
technews_df.columns = ['content', 'title', 'url','date'] #命名column name
technews_df = technews_df.drop(0)#刪去第一行

techorange_df = pd.DataFrame(techorange)
techorange_df = techorange_df.rename(columns = {"link": "url",})
dta = technews_df.append(techorange_df, sort=False)
dta.head()
```

### 段詞處理 (jieba)
雖然有三千多篇文章可以用來分析，然而因為分析時電腦一直當機，因此在這裡只先分析100篇

```python
result =[]
jieba.set_dictionary('dict.txt') #網路上下載的繁體辭典
for i in range(100):  #3868為總文章數
    w_lt = jieba.cut(dta.iloc[i,0],cut_all=True) #斷詞
    result1=[]
    for seg in w_lt : # 去除文中的空格與換行，並將之放入result1
        seg = ''.join(seg.split())
        if (seg != '' and seg != "\n" and seg != "\n\n") :
            result1.append(seg)
        content = ' '.join(result1) #重新合併同一段文章，斷詞中間用空格分開
    result.append(content)
```

```python
print(result[10]) #段詞結果
```

## TF-IDF 

```python
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = CountVectorizer()    
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(result))
```

```python
word = vectorizer.get_feature_names() #所有文本的關鍵字
weight = tfidf.toarray()              #對應的tfidf矩陣
print(word)
```

```python
#製作tfidf 矩陣
df_tfidf = pd.DataFrame(tfidf.toarray(),columns=word)
df_tfidf
```

```python
vectorizer1 = TfidfVectorizer()
X = vectorizer1.fit_transform(result)
indices = np.argsort(vectorizer1)[::-1]
features = vectorizer1.get_feature_names()
```

```python
for l in weight: 
    print ([(features[x],l[x]) for x in (l*-1).argsort()][:5]) #print 出每一篇文章中tf-idf前五高的詞與其tf-idf值
```
